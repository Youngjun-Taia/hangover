# 서울시 음주운전 사고 분석을 통한 부상(상해)수준 예측("음주운전 사고, 상해 정도를 미리 알 수 있을까? – 예측의 새로운 가능성")



- **기간:** 2024-09-05 ~ 2024-09-11 (총 5일)
- **정의:** 서울특별시의 10년 간 음주운전 사고 데이터를 통하여 음주운전 사고의 일정한 패턴을 분석하고, 이에 따른 부상정도를 예측하는 모델을 구성.
- **목적:**
  - 음주운전 사고의 주요 원인 분석 및 상해 수준과의 상관관계 도출
     - 음주운전 사고 발생 시 사고 유형, 운전자 특성 등이 부상 수준에 미치는 영향을 분석하여, 사고 발생에 따른 상해 수준을 예측한다.
  - 시간대 및 기상상태에 따른 음주운전 사고 발생 패턴 분석
     - 특정 시간대나 기상 상태에서 음주운전 사고가 빈번하게 발생하는 패턴을 분석하여, 해당 요인들이 상해 수준에 어떻게 영향을 미치는지 예측할 수 있는 모델을 개발한다.
  - 예방 및 정책 수립을 위한 음주운전 사고의 상해 예측 모델 개발
     - 머신러닝을 통해 사고 발생 시 상해 수준을 미리 예측하여, 정책 수립 및 교통사고 예방을 위한 가이드라인을 제공하며, 위험도가 높은 상황에서 사고 예방 대책을 마련한다.
- **범위:**
  - 서울특별시 내 모든 음주운전 사고에 대해서 조사한다.
  - 2023 ~ 2014년, 10년치 데이터를 수집한다.
  - 데이터 수집 후, 부상수준에 미치는 영향, 특정 상황에 따른 음주운전 사고 발생에 따른 패턴의 유무를 확인한다.
