```python
accident = pd.read_csv('accident.csv')
```

```python
accident.drop(['사고번호', '시군구'], axis= 1)
```

```python
accident.columns
```

```python
df = accident[['사망자수', '중상자수', '경상자수', '부상신고자수', '가해운전자 성별']]
```

```python
df.rename(columns={'가해운전자 성별': '가해운전자_성별'}, inplace=True)
```

```python
# MANOVA를 활용한 분석 실시.(다수의 수치형과 범주형을 비교 할 때 용이)
import statsmodels.api as sm
from statsmodels.formula.api import ols
from statsmodels.multivariate.manova import MANOVA
# OLS 모델 생성
manova = MANOVA.from_formula('사망자수 + 중상자수 + 경상자수 + 부상신고자수 ~ 가해운전자_성별', data=df)
manova_results = manova.mv_test()

# 결과 출력
print(manova_results)
```

```python
# 시각화 자료 비교, 일부 수치의 규모가 차이나서 로그스케일 적용
mean_scores = df.groupby('가해운전자_성별').mean().reset_index()

# 데이터프레임을 길게 변환
mean_scores_melted = mean_scores.melt(id_vars='가해운전자_성별', var_name='사상자 유형', value_name='평균 수')

# 막대 그래프 그리기
plt.figure(figsize=(12, 8))
sns.barplot(x='가해운전자_성별', y='평균 수', hue='사상자 유형', data=mean_scores_melted, palette='viridis')
plt.yscale('log')
plt.title('가해운전자 성별에 따른 사상자수 평균')
plt.xlabel('가해운전자_성별')
plt.ylabel('평균 수')
plt.legend(title='사상자 유형')
plt.show()
```

```python
# 시각화 자료 비교, 일부 수치의 규모가 차이나서 정규화 실시 [0~1]로 수정하는 것.
from sklearn.preprocessing import MinMaxScaler

# 데이터 정규화
scaler = MinMaxScaler()
mean_scores_scaled = mean_scores.copy()
mean_scores_scaled[['사망자수', '중상자수', '경상자수', '부상신고자수']] = scaler.fit_transform(mean_scores[['사망자수', '중상자수', '경상자수', '부상신고자수']])

# 데이터프레임을 길게 변환
mean_scores_scaled_melted = mean_scores_scaled.melt(id_vars='가해운전자_성별', var_name='사상자 유형', value_name='정규화된 평균 수')

# 막대 그래프 그리기
plt.figure(figsize=(12, 8))
sns.barplot(x='가해운전자_성별', y='정규화된 평균 수', hue='사상자 유형', data=mean_scores_scaled_melted, palette='viridis')
plt.title('가해운전자_성별 정규화된 사상자수 평균')
plt.xlabel('가해운전자_성별')
plt.ylabel('정규화된 평균 수')
plt.legend(title='사상자 유형')
plt.show()
```

```python
# 서브플롯 생성
fig, axes = plt.subplots(nrows=2, ncols=2, figsize=(15, 10))

# 각 변수에 대해 막대 그래프 그리기
variables = ['사망자수', '중상자수', '경상자수', '부상신고자수']
for i, var in enumerate(variables):
    row, col = divmod(i, 2)
    sns.barplot(x='가해운전자_성별', y=var, data=mean_scores, ax=axes[row, col], palette='viridis')
    axes[row, col].set_title(f'{var} by 가해운전자_성별')
    axes[row, col].set_xlabel('가해운전자_성별')
    axes[row, col].set_ylabel(var)

plt.tight_layout()
plt.show()
```

```python


# 범주형 변수를 더미 변수로 변환
df_dummies = pd.get_dummies(df, columns=['가해운전자_성별'], drop_first=True)

# 독립 변수와 종속 변수 정의
X = df_dummies[['중상자수', '경상자수', '부상신고자수', '가해운전자_성별_여']]
y = df_dummies['사망자수']

# 상수항 추가
X = sm.add_constant(X)

# 다중 회귀 모델 정의
model = sm.OLS(y, X).fit()

# 결과 출력
print(model.summary())
```
